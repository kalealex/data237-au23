```python
import pandas as pd

df = pd.read_csv("../data/iris.csv")

df.head()
```

```python
df["id"] = df.index

long_df = pd.wide_to_long(
    df,
    stubnames=["petal", "sepal"],
    sep="_",
    i="id",
    j="dimension",
    suffix=r'\w+').reset_index()

long_df.head()
```

## Vis from long df
...

```python
import altair as alt

alt.Chart(long_df).mark_circle().encode(
    x="sepal:Q",
    y="petal:Q"
)
```

```python
alt.Chart(long_df).mark_circle().encode(
    x="sepal:Q",
    y="petal:Q",
    column="dimension:N"
)
```

```python
alt.Chart(long_df).mark_circle().encode(
    x="sepal:Q",
    y="petal:Q",
    color="species:N",
    column="dimension:N"
)
```

Note the clustering of points by species in this coord space.
